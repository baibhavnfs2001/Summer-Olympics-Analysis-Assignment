### 2. Which sport is having most number of Gold Medals so far? (Top 5)

```python
df_gold = proc_df[proc_df.Medal == 'Gold']
```

```python
df_gold
```

```python
lsGold = []

for Sport in df_gold.Sport.unique():
    
    lsGold.append([Sport, len(df_gold[df_gold.Sport == Sport])])

lsGold = pd.DataFrame(lsGold, columns =['Sport','Gold Medals'])

lsGold = lsGold.sort_values(by = 'Gold Medals', ascending = False).head()

lsGold.plot(x = 'Sport', y = 'Gold Medals', kind = 'bar', figsize = (7,5))

print ("The Sport having most number of gold medals is : ", lsGold.Sport[0])
```

```python
lsGold
```
